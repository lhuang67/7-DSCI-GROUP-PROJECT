### Players dataset

- Variables:

  - `subscribe` (lgl): TRUE if the player subscribed to the newsletter, FALSE otherwise
  - `played_hours` (dbl): Total hours played by each player
  - `Age` (dbl): Age of the player
  - `hashedEmail`, `name`, `gender`: Player identifiers and demographic info
- Notes:
  - The `players` dataset contains 2 missing values in the `Age` column, while all other variables are complete. Since the proportion of missing data is very small, these rows can be ignored. 
  - Some variables (such as name or hashedEmail) are identifiers and are not useful for our purposes.

### Sessions dataset
- Observations: 1535 sessions  
- Variables: 5  
- Variables:

  - `start_time`, `end_time`: Times of play session
  - `original_start_time`, `original_end_time`: Timestamps for processing
- Notes:
  - Useful for analysis of player activity, each row represents one play session.

### Question  
- Broad question: Question 1
- Specific Question: Can a player's total played hours, age, experience level and gender, predict whether they subscribe to the game-related newsletter?

### Variables  
- **Target variable:**  
  - `subscribe`  


- **Predictors variables:**  
  - `played_hours`
  - `Age`
  - `experience`
  - `gender`

### Why this question?  
This question explores which types of players are most engaged in the community.
All required predictors and the target variable are available in `players.csv`.


```r
# Put the dataset in tidy format
players_tidy <- players |>
  select(-hashedEmail, -name) |>
  mutate(experience = as.factor(experience), gender = as.factor(gender))

head(players_tidy)
```

```r
# Mean value for each quantitative variable
summarise(players_tidy,  mean_age = mean(Age, na.rm = TRUE), mean_hours = mean(played_hours, na.rm = TRUE))
```

```r
# Exploratory Visualizations

# Subscription status vs Experience Level
ph_vs_ns <- ggplot(players_tidy, aes(x = experience, fill = subscribe)) +
  geom_bar(position = "dodge") +
  labs(
    title = "Subscription Status by Experience Level",
    x = "Experience Level",
    y = "Number of Players",
    fill = "Subscribed"
  )
ph_vs_ns

# Distribution of Total Played Hours
total_hours_plot <- ggplot(players_tidy, aes(x = played_hours)) +
  geom_histogram(color = "white") +
  labs(
    title = "Distribution of Total Played Hours",
    x = "Total Played Hours (hours)",
    y = "Number of Players"
  )
total_hours_plot

# Played Hours vs Age by Subscription Status
player_age_vs_age <- ggplot(players_tidy, aes(x = Age, y = played_hours, color = subscribe)) +
  geom_point(alpha = 0.7) +
  labs(
    title = "Played Hours vs Age by Subscription Status",
    x = "Player Age (years)",
    y = "Total Played Hours (hours)",
    color = "Subscribed"
  ) 
player_age_vs_age

# Subscription Status by Gender
gender_vs_subscribe <- ggplot(players_tidy, aes(x = gender, fill = subscribe)) +
  geom_bar(position = "dodge") +
  labs(
    title = "Subscription Status by Gender",
    x = "Gender",
    y = "Number of Players",
    fill = "Subscribed"
  ) +
  theme_minimal()

gender_vs_subscribe
```

### Overall Insights from the Exploratory Visualizations

Only a small subset of highly active players spends many hours playing the game and is also more likely to subscribe to the newsletter. Experience appears to be positively correlated with subscription, while age shows no clear correlation with playing time or subscription and may not be a strong predictor. Male players make up the largest share of the dataset and also represent the majority of newsletter subscribers.

## Methods and Plan

### Proposed Method
To address the my question, I plan to use a **classification model**.

### Why this method is appropriate
Since the response variable `subscribe` is categorical, a classification approach is appropriate.   
The model will use both the predictors to estimate the probability of a player subscribing to the newsletter.

### Assumptions
The main assumptions for classification methods are:
- Each observation (player) is independent of the others.
- The predictors have a significant relationship with the target variable (subscribe).
- The data is sufficiently balanced between the two classes (TRUE/FALSE) to avoid bias in the predictions.

These assumptions will be checked before training the model.

### Potential limitations
Classification models can be sensitive to **unbalanced data**, where one class significantly outnumbers the other.
They can also be influenced by irrelevant or repetitive predictors, so careful selection of these is important.

### Model comparison and selection
Model performance will be evaluated using metrics such as accuracy, precision, and recall on the test dataset.
Cross-validation will be used to ensure the model is well calibrated to previously unseen data.

### Data processing plan
The variables (experience, gender) will be converted into factors so they can be used in the classification model.
The dataset will be split into:
- Training set (75%), used to fit the model
- Testing set (25%), used to evaluate the model's performance

This split will be performed before any modeling using the `initial_split()` function.
**Cross-validation** will be applied within the training set to ensure model stability and prevent overfitting and underfitting.
